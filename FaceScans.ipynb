```python
import os
from zipfile import ZipFile
import pandas as pd
from deepface import DeepFace
```

```python
# Process all images in the faceimages folder
image_folder = "C:/Users/frost/Desktop/deepface/faceimages"
```

```python
image_files = [f for f in os.listdir(image_folder) if f.endswith(('.jpg', '.jpeg', '.png'))]
```

```python
# Create a DataFrame to store results
results_df = pd.DataFrame(columns=["Filename", "Gender", "Race/ethnicity", "Age"])
```

```python
# Process each image and fill the DataFrame
for filename in image_files:
    image_path = os.path.join(image_folder, filename)
    
    try:
        # Use deepface to analyze the image
        result = DeepFace.analyze(img_path=image_path, actions=['gender', 'race', 'age'])
        
        # Append the results to the DataFrame
        results_df = results_df.append({
            "Filename": filename,
            "Gender": result["gender"],
            "Race/ethnicity": result["dominant_race"],
            "Age": result["age"],
        }, ignore_index=True)
        
    except Exception as e:
        # Handle exceptions (e.g., if deepface cannot analyze the image)
        results_df = results_df.append({
            "Filename": filename,
            "Gender": "unknown",
            "Race/ethnicity": "unknown",
            "Age": -1,
        }, ignore_index=True)
```

```python
# Write the results to a CSV file
csv_output_path = "C:/Users/frost/Desktop/deepface/FacesScanned.csv"
results_df.to_csv(csv_output_path, index=False)
```
